```python
import pandas as pd

census_df = pd.read_csv('https://storage.googleapis.com/py_ml_datasets/census.csv')
print(census_df.head())
census_df.shape
```

```python
#Q1:哪個州(state)的郡(county)數最多

def answer_one(df):    
    return census_df['STNAME'].value_counts().idxmax()

print(answer_one(census_df))
```

```python
#Q2:僅考慮每州(state)人口最多的三個郡(county)，計算人口總合(CENSUS2010POP)，哪三個州總和數最多(請注意SUMLEV變數)
#SUMLEV
#思路 先把每個州人口最多的三個郡篩選出來，後把CENSUS2010POP加總，然後回傳最多的
```

```python
census_df['SUMLEV']
```

```python
census_df['SUMLEV'].value_counts()

#40的是state data 不是county data 所以要先排除掉
```

```python
county_data = census_df[census_df['SUMLEV'] == 50] #篩選所有SUMLEV等於50的COUNTY
state_names = county_data['STNAME'].unique() #51個名稱
print(state_names) #拿到一個所有STATE name的list
```

```python
#有51個STATE 所以要跑這51個STATE裡面所有的COUNTY的CENSUS2010POP的值然後加總起來

top3_pop_summations = []
for state_name in state_names:
    single_state = county_data[county_data['STNAME'] == state_name] #遞迴到哪一個就把那個stname下的所有資料拿出來 叫做 single_state
    single_state_sorted = single_state.sort_values('CENSUS2010POP', ascending=False) #把這所有資料做排序，依照CENSUS2010POP這個欄位，降冪排序
    top3_pop = single_state_sorted['CENSUS2010POP'].values[:3] #此時single_state_sorted['CENSUS2010POP']會有全部county的資料 排序後取前3 
    #這邊的top3_pop 是一個series 由一個index跟value組成 所以取value[:3]會是前三名的值
    top3_pop_sum = top3_pop.sum()
    top3_pop_summations.append(top3_pop_sum)
print(top3_pop_summations) #得到前三多人口數總和加總list 長度為51
```

```python
ser = pd.Series(top3_pop_summations, index =state_names) #把兩個list變成一個series
print(ser.sort_values(ascending=False))
print('-'*50)
print(ser.sort_values(ascending=False)[:3])
print('-'*50)
print(*list(ser.sort_values(ascending=False)[:3].index))
```

## 第二種解法:使用nlargest
傳回指定變數前n大的值跟索引出來

```python
#第二種解法 使用nlargest 傳回指定變數前n大的值跟索引出來

census_df['CENSUS2010POP'].nlargest(3) #在這三個列裡面 有三個最大數字如右
```

```python
#做第一次groupby 把相同州名稱的groupby起來
groupby_object = county_data.groupby('STNAME')
print(groupby_object)
#要對哪一個變數做什麼摘要 對人口數這個欄位取前三大
groupby_object['CENSUS2010POP'].nlargest(3) 
```

```python
top3_by_states = groupby_object['CENSUS2010POP'].nlargest(3)

#再用一次groupby把stname組合起來
groupby_object = top3_by_states.groupby('STNAME')
groupby_object.sum().nlargest(3).index
#multiindex的索引
```

```python
def answer_two(df):
    # filter county_data
    county_data = df[df['SUMLEV'] == 50]
    # nlargest at state level
    groupby_object = county_data.groupby('STNAME')
    top3_by_states = groupby_object['CENSUS2010POP'].nlargest(3)
    groupby_object = top3_by_states.groupby('STNAME')
    # nlargest at nation level
    ser = groupby_object.sum()
    ans = list(ser.nlargest(3).index)
    return ans

print(*answer_two(census_df))
```

```python
# Q3 哪個郡(county) 在2010-2015期間人口改變數量最高 (POPESTIMATE2010:POPESTIMATE2015)這六個變數

#解題想法：延續上一提，遮罩剩下顯示county的 把SUMLEV=40的遮掉不看
#然後新增一個欄位計算POPESTIMATE2010:POPESTIMATE2015裡面最大的-最小的就是答案

county_data = census_df[census_df['SUMLEV'] == 50]

# ct_list = list(census_df['CTYNAME'].unique())
# len(ct_list) #有1927個county

county_data.loc[:,'POPESTIMATE2010':'POPESTIMATE2015']
```

```python
pop_max = county_data.loc[:,'POPESTIMATE2010':'POPESTIMATE2015'].max(axis=1)
pop_min = county_data.loc[:,'POPESTIMATE2010':'POPESTIMATE2015'].min(axis=1)
pop_diff = pop_max - pop_min
pop_diff.idxmax()
```

```python
def answer_three(df):
    # filter county_data
    county_data = df[df['SUMLEV'] == 50]
    pop_max = county_data.loc[:,'POPESTIMATE2010':'POPESTIMATE2015'].max(axis=1)
    pop_min = county_data.loc[:,'POPESTIMATE2010':'POPESTIMATE2015'].min(axis=1)
    pop_diff = pop_max - pop_min
    max_pop_diff_index = pop_diff.idxmax()
    ans = county_data['CTYNAME'][max_pop_diff_index]
    return ans

print(answer_three(census_df))
print(county_data.loc[2667,['STNAME','CTYNAME']])
```
