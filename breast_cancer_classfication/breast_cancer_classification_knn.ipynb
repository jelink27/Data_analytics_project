# **關聯分析 (correlation map)**

```python
# correlation calculate
corr = df[['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error',
       'fractal dimension error', 'worst radius', 'worst texture',
       'worst perimeter', 'worst area', 'worst smoothness',
       'worst compactness', 'worst concavity', 'worst concave points',
       'worst symmetry', 'worst fractal dimension', 'target']].corr()
plt.figure(figsize=(40,40))
sns.heatmap(corr, square=True, annot=True, cmap="RdBu_r") #center=0, cmap="YlGnBu"
```

```python
# Calculate the correlations between the features and the label
corr = df.corr()['target'].sort_values(ascending=True)

# 找出關聯性最高的六個 (取絕對值)
corr_top_6 = corr.abs().sort_values(ascending=False)[:6] 

print(corr_top_6)
# Get the values for the top six features
```

# **切割訓練集與測試集**

```python
from sklearn.model_selection import train_test_split
X = df[['worst concave points', 'worst perimeter', 'mean concave points', 'worst radius', 'mean perimeter']]
y = df['target'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 27)
```

```python
print('Training data shape:',X_train.shape)
print('Testing data shape:',X_test.shape)
```

## 建立 k-nearest neighbors(KNN) 模型
Parameters:
- n_neighbors: 設定鄰居的數量(k)，選取最近的k個點，預設為5。
- algorithm: 搜尋數演算法{'auto'，'ball_tree'，'kd_tree'，'brute'}，可選。
- metric: 計算距離的方式，預設為歐幾里得距離。

Attributes:
- classes_: 取得類別陣列。
- effective_metric_: 取得計算距離的公式。

Methods:
- fit: 放入X、y進行模型擬合。
- predict: 預測並回傳預測類別。
- score: 預測成功的比例。

```python
from sklearn.neighbors import KNeighborsClassifier

# 建立KNN模型
knnModel = KNeighborsClassifier(n_neighbors=3)
# 使用訓練資料訓練模型
knnModel.fit(X_train,y_train)
# 使用訓練資料預測分類
predicted= knnModel.predict(X_train)
```

# **使用Score評估模型**

```python
# 預測成功的比例
print('訓練集: ',knnModel.score(X_train,y_train))
print('測試集: ',knnModel.score(X_test,y_test))
```

# **使用混淆矩陣**

```python
from sklearn.metrics import confusion_matrix
y_pred = knnModel.predict(X_test)
confusion_mat = confusion_matrix(y_test, y_pred)
print(confusion_mat)
```
